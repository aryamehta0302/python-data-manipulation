This notebook presents a detailed analysis of the salary dataset provided. The goal is to extract meaningful insights and trends based on job title, experience, company location, and more.

## üîç Step 1: Importing Libraries

We begin by importing the necessary libraries.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set(style='whitegrid')
```

## üì• Step 2: Loading the Dataset

```python
df = pd.read_csv("salaries.csv")
df.head()
```

## üßæ Step 3: Data Overview

```python
df.info()
```

## üîç Step 4: Checking for Missing Values

```python
df.isnull().sum()
```

## üìä Step 5: Summary Statistics

```python
df.describe(include='all')
```

## üíµ Step 6: Salary Distribution (USD)

```python
plt.figure(figsize=(10,5))
sns.histplot(df['salary_in_usd'], bins=50, kde=True)
plt.title('Salary Distribution in USD')
plt.xlabel('Salary (USD)')
plt.ylabel('Frequency')
plt.show()
```

## üìà Step 7: Average Salary by Experience Level

```python
exp_salary = df.groupby("experience_level")["salary_in_usd"].mean().sort_values()
plt.figure(figsize=(8,4))
sns.barplot(x=exp_salary.index, y=exp_salary.values, palette="viridis")
plt.title("Average Salary by Experience Level")
plt.ylabel("Average Salary (USD)")
plt.xlabel("Experience Level")
plt.show()
```

## üßë‚Äçüíº Step 8: Top Paying Job Titles

```python
top_jobs = df.groupby("job_title")["salary_in_usd"].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=top_jobs.values, y=top_jobs.index, palette="mako")
plt.title("Top 10 Highest Paying Job Titles")
plt.xlabel("Average Salary (USD)")
plt.ylabel("Job Title")
plt.show()
```

## üè† Step 9: Remote Work and Salary

```python
remote_avg_salary = df.groupby("remote_ratio")["salary_in_usd"].mean()
plt.figure(figsize=(7,4))
sns.barplot(x=remote_avg_salary.index.astype(str), y=remote_avg_salary.values, palette="coolwarm")
plt.title("Average Salary by Remote Work Ratio")
plt.xlabel("Remote Work %")
plt.ylabel("Average Salary (USD)")
plt.show()
```

## üåç Step 10: Company Location vs Salary

```python
top_locations = df.groupby("company_location")["salary_in_usd"].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_locations.index, y=top_locations.values, palette="cubehelix")
plt.title("Top Paying Company Locations")
plt.xlabel("Company Location")
plt.ylabel("Average Salary (USD)")
plt.show()
```

## üìå Final Observations and Conclusion

- üë®‚Äçüî¨ Senior-level professionals earn the highest average salary, followed by mid-level.
- üíº Certain job roles like **Data Architect** and **ML Engineer** offer significantly higher pay.
- üè† Remote jobs (100% remote) show competitive salaries and are becoming more common.
- üåé Company location plays a vital role‚Äîcompanies based in certain countries like the US offer higher average salaries.

This analysis highlights the importance of experience, remote work flexibility, and job specialization in determining salary. Job seekers can align their goals accordingly to maximize earnings potential.
